```python
s = ''
grand_total = 0
def billing_system():
    global s,grand_total
    n = int(input("Enter n = "))
    for i in range(n):
        name  = input("Enter name = ")
        price = int(input("Enter price = "))
        quantity = int(input("Enter quantity = "))
        total = price * quantity

        info = name+" "+str(price)+" "+str(quantity) + " "+str(total)
        s = s+info+"\n"
        grand_total = grand_total +total
    print(s)
    print(grand_total)
    x = input("Enter y for more calculation")
    if x == 'y':
        billing_system()

billing_system()
```

```python
def add(a,b):
    return a+b
def sub(a,b):
    return a+b
def mul(a,b):
    return a*b
def div(a,b):
    return a/b
def cal():
    a = int(input("Enter a = "))
    b = int(input("Enter b = "))
    o = input("Enter + - * /  ")
    if o == "+":
        print(add(a,b))
    elif o == '-':
        print(sub(a,b))
    elif o == '*':
        print(mul(a,b))
    elif o == '/' and b!=0:
        print(div(a,b))
    elif o == '/' and b==0:
        print("The value of b can not be zero.")
    else:
        print("Invalid operator")
        
    x = input("Enter y for more cal ")
    if x == 'y':
        cal()
cal()
```
