```python
import csv
csv_columns = ['Name', 'Age', 'Address']
dicts = [{'Name': 'Ram', 'Age': '34', 'Address': 'Kathmandu'},
        {'Name': 'Shyam', 'Age': '54', 'Address': 'Bara'},
        {'Name': 'Nabin', 'Age': '45', 'Address': 'Sauhara'}]

b = open("data.csv", 'w')

writer = csv.DictWriter(b, fieldnames=csv_columns)
writer.writeheader()

for data in dicts:
    writer.writerow(data)
```
